**Punto 3: Sacar negativo de la imagen**

```
def negative(l):
  img =cv2.imread('cafe.jpg',cv2.IMREAD_GRAYSCALE)
  row, column = img.shape
  img2 = np.zeros((row, column), dtype= np.uint8)
  for x in range(row):
    for y in range(column):
      img2[x,y]= (l-1)- img[x,y]
    
  displayGray2(img,img2,'imagenA','ImagenB')
  return

negative(256)
```
